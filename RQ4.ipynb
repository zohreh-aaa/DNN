```python
############################################################## Call_functions  ##########################################
import numpy as np
import matplotlib.pyplot as plt
data_name=input("Enter the dataset           mnist/ cifar10    : ")
model_name=input("Enter the model name    LeNet5  /  12Conv     : ")
print(data_name, model_name)
```

```python
if data_name=="mnist":
  path="/content/drive/MyDrive/Paper/Codes/RQ2_3/Correlation/BB_mnist_BB_LeNet5/Size"
  data_name="mnist"
  model_name="LeNet5"
  det_name="timedet_"
if data_name=="cifar10": 
  path="/content/drive/MyDrive/Paper/Codes/RQ2_3/Correlation/BB_cifar10_ResNet20/"
  data_name="cifar10"
  model_name="ResNet20"
  det_name="timegd_"
ld=[]
LSC=[]
DSC=[]
kmnc=[]
nbc=[]
mis=[]
snac=[]
tknc=[]
nc=[]
std=[]

for size in [100,200,300,400]:
#timedet or timegd chang1 path
  ld.append(np.load(path+"metrics/"+det_name+ str(size)+".npy"))
  LSC.append(np.load(path+"metrics/timelsc_"+ str(size)+".npy"))
  DSC.append(np.load(path+"metrics/timedsc_"+ str(size)+".npy"))
  kmnc.append(np.load(path+"metrics/timekmnc_"+ str(size)+".npy"))
  nbc.append(np.load(path+"metrics/nbc_"+ str(size)+".npy"))
  snac.append(np.load(path+"metrics/snac_"+ str(size)+".npy"))
  tknc.append(np.load(path+"metrics/timetknc_"+ str(size)+".npy"))
  nc.append(np.load(path+"metrics/timenc_"+ str(size)+".npy"))
```

```python
TPP0=np.load("/content/drive/MyDrive/Paper/Codes/RQ4/BB_MNIST_internal_layer/60_VGG_PPtime_Size100.npy")
TPP1=np.load("/content/drive/MyDrive/Paper/Codes/RQ4/BB_MNIST_internal_layer/60_VGG_PPtime_Size200.npy")
TPP2=np.load("/content/drive/MyDrive/Paper/Codes/RQ4/BB_MNIST_internal_layer/60_VGG_PPtime_Size300.npy")
TPP3=np.load("/content/drive/MyDrive/Paper/Codes/RQ4/BB_MNIST_internal_layer/60_VGG_PPtime_Size400.npy")
```

```python
ld[0]=TPP0+ld[0]
ld[1]=TPP1+ld[1]
ld[2]=TPP2+ld[2]
ld[3]=TPP3+ld[3]
```

```python
for i, j in zip([ld, LSC, DSC, nc, tknc, kmnc], ["GD", "LSC", "DSC", "NC", "TKNC", "KMNC"]):
  plt.boxplot(i)
  plt.xticks([1,2,3,4], [100,200,300,400])
  plt.xlabel("Subset size")
  plt.ylabel("Execution time (s)")
  plt.title(str(j))
  plt.show()

```

```python

ld1=[ld[i].mean() for i in range(len(ld))]
DSC1=[DSC[i].mean() for i in range(len(DSC))]
LSC1=[LSC[i].mean() for i in range(len(LSC))]
nc1=[nc[i].mean() for i in range(len(nc))]
kmnc1=[kmnc[i].mean() for i in range(len(kmnc))]
tknc1=[tknc[i].mean() for i in range(len(tknc))]
```

```python
import matplotlib.pyplot as plt
import numpy as np
  
x=["100","200","300","400"]
# first plot with X and Y data
plt.plot(x, kmnc1, "--bo", color="brown")

plt.plot(x, DSC1,'-.', marker = '*', color="purple")
plt.plot(x, LSC1, '-.', marker = 'o', color="blue")
plt.plot(x, tknc1, '-.', marker = 's', color="grey")
plt.plot(x, nc1, '-.', marker = 'x', color="orange")
# plt.plot(x, std1, '-.', marker = '*', color="black")
plt.plot(x, ld1, '-.', marker = 'h',color="green")


plt.xlabel("Subset size")
plt.ylabel("Average execution time (s)")
# plt.title('Cifar10 / ResNet20')
# "TKNC", "NC",
#"KMNC","DSC","LSC","TKNC", "NC",
plt.legend(["KMNC","DSC","LSC","TKNC", "NC","GD"], loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
